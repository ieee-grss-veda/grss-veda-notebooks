## Workflow

1. Connect to the GRSS VEDA STAC catalog using `pystac_client`
2. Enumerate all available collections and display their titles

## Connect to the STAC Catalog

```python
from pystac_client import Client

STAC_API_URL = "https://api.dev.veda.grss.cloud/stac"
catalog = Client.open(STAC_API_URL)
```

## Display All Collections

```python
collections = list(catalog.get_collections())
print(f"Found {len(collections)} collections:\n")

for collection in sorted(collections, key=lambda x: x.title or x.id):
    print(f"- {collection.title or collection.id}")
```
